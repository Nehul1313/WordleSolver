```python
#Load the application
from nltk.corpus import words

#Global Dictionary of possible words
wordlist = list(words.words())
invalid_words = {'manid','sancy','gluck','cunas','lucan','lanka','kusan','ribat','belis','gareh','zymin'
,'dasnt','trink','pouce','aotus','waily','latro','melch','skair','stary','haire','latus','albus','vince',
'igdyr'}


dictionary = set()
for word in wordlist:
    if len(word)==5 and word not in invalid_words:
        dictionary.add(word.lower())
#The word means was not included in nltk
#Added the word manually
dictionary.add('means')  
saved_dictionary = dictionary.copy()
dictionary = list(dictionary)
dictionary.sort(reverse=True,key=lambda x: (len(set(x))) )

#Initializing state of the characters
unmatched_chars = set() 
matched_false_word = ""
matched_false_position = {}
matched_true_position = {}

#Function that updates character_state_variables: unmatched_chars, matched_false_position, matched_true_position
#from given parameters: word,status 

def mapWordle(status,word):
    if word=='':
        return
    
    global unmatched_chars, matched_false_word
    global  matched_false_position, matched_true_position
    unmatched_chars = set() 
    matched_false_word = ""
    matched_false_position = {}
    matched_true_position = {}
        
    for ind, char in enumerate(word):
        if status[ind] ==0:
            unmatched_chars.add(char)
        elif status[ind]==1:
            matched_false_position[char] = ind
            matched_false_word += char
        else:
            matched_true_position[ind] = char 

#Function that updates the dictionary by eliminating redundancies
#returns a list of top 5 words

def SolveWordle(status,word=''):
    global dictionary
    mapWordle(status,word)
    temp = set()
    for word in dictionary:
        flag = True
        for index,char in enumerate(word):
            if char in unmatched_chars:
                flag = False
                break
            #all characters must be present in the word
            if char in matched_false_position and matched_false_position[char]==index:
                flag = False
                break
            if index in matched_true_position and matched_true_position[index]!=char:
                flag = False
                break
        if flag:
            temp.add(word)

    dictionary = list(temp)
    dictionary.sort(reverse=True,key=lambda x: len(set(x)))
    #print(dictionary[0])

    from collections import Counter
    temp = set()
    for word in dictionary:
        if (Counter(word) & Counter(matched_false_word) == Counter(matched_false_word)):
            temp.add(word)

    dictionary = list(temp)
    dictionary.sort(reverse=True,key=lambda x: len(set(x)))
    return dictionary[0:5]

from IPython.display import clear_output

match = False
true_status = [2,2,2,2,2]
trial = 1
status_string = 'Status format:unmatched:0, incorrect position:1, correct position:2\n'  
```

```python
while not match:
    if trial==1:
        word =''
        status = [0,0,0,0,0]
    else:
        word = input('choice of word: ')
        status = input(status_string)
        status = list(status)
        status = list(map(int,status))
        clear_output(wait=True)
    
    possible_words = SolveWordle(status,word)
    print('Trial No: ',trial)
    print('Total possibilities: ', len(dictionary))
    print('top 5 available choices:\n',possible_words,)
    
    match = status==true_status
    trial+=1
    
print('Yay! we have successfully found the word.')
dictionary = saved_dictionary
```
